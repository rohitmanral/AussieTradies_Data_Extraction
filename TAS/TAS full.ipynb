### Getting headers

```python
page_source = driver.page_source
soup =  BeautifulSoup(page_source, 'lxml')
            
div_container = soup.find('tr', class_='GridHeader')  

headers=[]
for ptag in div_container.find_all('td'):
    # prints the p tag content
    headers.append(ptag.text)
    
headers
```

### Getting Data

```python
licensee_detail=[]

for i in range(1, int(pages)+1):
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    data = []
    for nested_soup in soup.find_all('tr', {'class':['GridItem', 'GridItemAlternating']}):
        for i in nested_soup.find_all('td'):
            data.append(i.text.strip())
    #data
    licensee_detail = licensee_detail + data
    
    driver.find_element_by_xpath('//*[@id="ctl00_ctl00_ctl00_ctlMainContent_ctlMainContent_MainContent_ctlOnlineSearch_ctlDataGridPager_btnNextPage"]').click()
    time.sleep(6)
    
licensee_detail
```

```python
len(licensee_detail)
```

```python
final_data=[licensee_detail[i:i + 7] for i in range(0, len(licensee_detail), 7)]
final_data
```

```python
df = pd.DataFrame(final_data, columns = headers)
df
```

## Land Surveyors Board TAS

https://surveyor.dpipwe.tas.gov.au/ext/

```python
driver.get('https://surveyor.dpipwe.tas.gov.au/ext/')
time.sleep(1)
```

```python
page_source = driver.page_source
soup =  BeautifulSoup(page_source, 'lxml')

data = [i.text for i in soup.findAll('td', class_='noWrap')]
#data
final_data=[data[i:i + 3] for i in range(0, len(data), 3)]
final_data
```

```python
df_land_sur = pd.DataFrame(final_data, columns = ['Last Name','First Name','Street address'])
df_land_sur.head()
```

```python
df_land_sur.shape
```

```python
xpaths=[]
for j in range(2,74): #creating xpaths: just change span[i] to click different links
    xpaths.append('//*[@id="main"]/table/tbody/tr['+str(j) + ']/td[1]/a')
                     
#print(xpaths) 
```

```python
len(xpaths)
```

```python
lic_nums=[]
adds=[]
phones=[]
emails=[]
websites=[]
exps=[]
comps=[]
orgs=[]





for i in range(1, len(xpaths)+1):
    
    driver.find_element_by_xpath(xpaths[i-1]).click()
    
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')
    
    
            
    lic_num=driver.find_element_by_id('surveyorNumber').text
    lic_nums.append(lic_num)
    
    info=driver.find_element_by_id('principalAddress').text
    #info
    info_n= info.split('\n')
    
    add= info_n[0].strip() + ', ' + info_n[1].strip() + ', ' + info_n[2].strip()
    adds.append(add)
    
    
    
    
    
    c_phn=0
    c_eml=0
    c_web=0

    for s in info_n[3:]:
    
        if "Phone" in s:
            phones.append((s.split(':'))[1].strip())
            c_phn=c_phn+1
        
        elif "Email" in s:
            emails.append((s.split(':'))[1].strip())
            c_eml=c_eml+1
        
        elif "Website" in s:
            websites.append((s.split(' '))[1].strip())
            c_web=c_web+1

    if c_phn==0:
        phones.append(' ')
    if c_eml==0:
        emails.append(' ')
    if c_web==0:
        websites.append(' ')
    
    
    
    exp=driver.find_element_by_xpath('//*[@id="main"]/div[3]/div[3]/div/table/tbody/tr[2]/td[3]').text
    exps.append(exp)
    
    comp= driver.find_element_by_xpath('//*[@id="main"]/div[3]/div[4]/div/table/tbody/tr[2]/td[1]').text
    comps.append(comp)
    
    org= driver.find_element_by_xpath('//*[@id="main"]/div[3]/div[4]/div/table/tbody/tr[2]/td[3]').text
    orgs.append(org)
    
    

    
    
    
    driver.get('https://surveyor.dpipwe.tas.gov.au/ext/')
    time.sleep(1)
```

### Add lists to existing dataframe

```python
df_land_sur['License number'] = lic_nums
df_land_sur['Business address'] = adds
df_land_sur['Phone'] = phones
df_land_sur['Email'] = emails
df_land_sur['Website'] = websites
df_land_sur['Expiry date'] = exps
df_land_sur['Specialist competencies'] = comps
df_land_sur['Accrediting organisation'] = orgs
```
