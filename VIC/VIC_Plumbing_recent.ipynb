# Main Program

```python
driver.get('https://consumer.etoolbox.pic.vic.gov.au/_layouts/cc/pic_validatepp.aspx')
```

```python
all_licensee_names=[]
all_licensee_detail=[]

for i in Letters_comb:
    driver.get('https://consumer.etoolbox.pic.vic.gov.au/_layouts/cc/pic_validatepp.aspx')
    
    driver.find_element_by_xpath('//*[@id="ctl00_PlaceHolderMain_uxSurname"]').send_keys(i)
    driver.find_element_by_xpath('//*[@id="ctl00_PlaceHolderMain_uxSearch"]').click()
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    licensee_names=[] 
    licensee_names = [i.text for i in soup.findAll('span', {'class': 'prac-result'})] #Getting the count of presesnt licenses for each letter combination
    #print(licensee_names)
    #len(licensee_names)
    all_licensee_names=all_licensee_names+licensee_names
    #print(all_licensee_names)
    
    xpaths=[]
    for j in range(1,len(licensee_names)+1): #creating xpaths: just change span[i] to click different links
        xpaths.append('//*[@id="aspnetForm"]/div[3]/table/tbody/tr[1]/td[2]/table/tbody/tr/td[1]/table/tbody/tr[5]/td/div/span['+str(j) + ']')
    #print(xpaths)  
    
    licensee_detail=[]
    for k in range(1,len(licensee_names)+1): #getting the details of every license in a combination
        try:
            driver.find_element_by_xpath(xpaths[k-1]).click()
            time.sleep(2)
    
            page_source = driver.page_source
            soup =  BeautifulSoup(page_source, 'lxml')
            
    
            licensee_detail.append([i.text for i in soup.findAll('table', {'id': 'tblRegSummary'})]) 
        
        except:
            print("No data for {}".format(i))
    
    #print(licensee_detail)  
    #len(licensee_detail)     
    all_licensee_detail=all_licensee_detail+licensee_detail
    #print(all_licensee_detail)
```
